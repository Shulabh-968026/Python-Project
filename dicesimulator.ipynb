# This is Dice Simulator

```python
import random 
print("---This is dice simulator program---")
x="y"
while("y"==x):
    n=random.randrange(1,7)
    if n==1:
        print("----------")
        print("|        |")
        print("|   0    |")
        print("|        |")
        print("----------")
    elif n==2:
        print("----------")
        print("|0       |")
        print("|        |")
        print("|       0|")
        print("----------")
    elif n==3:
        print("----------")
        print("|   0    |")
        print("|   0    |")
        print("|   0    |")
        print("----------")
    elif n==4:
        print("----------")
        print("|0      0|")
        print("|        |")
        print("|0      0|")
        print("----------")
    elif n==5:
        print("----------")
        print("|0      0|")
        print("|   0    |")
        print("|0      0|")
        print("----------")
    elif n==6:
        print("----------")
        print("|  0  0  |")
        print("|  0  0  |")
        print("|  0  0  |")
        print("----------")    
    x=input("Enter y for roll again::").lower()
```
